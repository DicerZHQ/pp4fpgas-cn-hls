# Import Overlay
导入Overlay

```python
import pynq.lib.dma
import numpy as np

sumol = pynq.Overlay("sum.bit")

# dma = overlay.const_multiply.multiply_dma
# multiply = overlay.const_multiply.multiply

dma = sumol.axi_dma_0
# s = sumol.sum_0
```

```python
from pynq import Xlnk

xlnk = Xlnk()
in_buffer = xlnk.cma_array(shape=(128,), dtype=np.int32)
out_buffer = xlnk.cma_array(shape=(128,), dtype=np.int32)

for i in range(128):
    in_buffer[i] = 3;

dma.sendchannel.transfer(in_buffer)
dma.recvchannel.transfer(out_buffer)
# dma.sendchannel.wait()
# dma.recvchannel.wait()

print(in_buffer)
print(out_buffer)
```
